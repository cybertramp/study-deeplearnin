```
# 1. 필요한 패키지 임포트
import tensorflow as tf

# 2. 데이터 셋 생성하기
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.reshape(60000, 784).astype('float32') / 255.0
x_test = x_test.reshape(10000, 784).astype('float32') / 255.0

y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)

# 3. 모델 구성하기
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=64, input_dim=28*28, activation='relu'))
model.add(tf.keras.layers.Dense(units=10, activation='softmax'))

# 4. 모델 학습 알고리즘 지정
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

# 5. 학습
hist = model.fit(x_train, y_train, epochs=5, batch_size=32)

# 6. 학습 history 살펴보기
print('loss =', hist.history['loss'])
print('accuracy =', hist.history['acc'])

# 7. 학습된 모델 평가하기
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss and metrics = ', loss_and_metrics)

# 8. 학습된 모델 사용하기 
y = model.predict(x_test[0:1])
print('prediction = ', y) # prediction의 원소들의 합은 1.0
```
